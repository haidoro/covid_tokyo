## jsonデータを読み込み

```
import pandas as pd
import json
import numpy as np


tokyo_data = pd.read_json('/content/drive/My Drive/AI/授業/covid/tokyo_data.json')
```

```
tokyo_data['patients']['data']
```

## DataFrame化

```
df_tokyo = pd.DataFrame( tokyo_data['patients']['data'])
df_tokyo
```

## 症状別累計人数

```
total = len(df_tokyo)
d = (df_tokyo["備考"].str.contains('死亡')).sum()
se = (df_tokyo["備考"].str.contains('重症')).sum()
si = (df_tokyo["備考"] == "中等症").sum()
mi = df_tokyo["備考"].isnull().sum()
di = (df_tokyo["退院"] == "〇").sum()
print('東京都の新型コロナウイルス感染者はトータル:',total,'死亡:',d,'重症:',se,'中症:',si,'軽傷:',mi,'退院:',di)
print('死亡確率：{}%'.format((d/total)*100),'退院率：{}%'.format((di/total)*100))
```
